```r
# Initialize a list to store the header content for each athlete
hea_contents <- list()

# Loop through all 28 .hea files
for (i in 1:28) {
  # Generate the file path based on the athlete number
  # Assuming file names are consistently formatted
  file_path <- sprintf("/content/ath_%03d.hea", i)

  # Read the .hea file content
  content <- readLines(file_path)

  # Store the content in the list
  hea_contents[[i]] <- content

  # Optionally, print the content or just the file name to confirm each file is processed
  print(paste("Content of", file_path, ":"))
  print(content)
}

# hea_contents now contains the header information for all 28 athletes
```

```r
# Initialize a list to store the ECG data for each athlete
ecg_data_list <- list()

# Number of athletes
num_athletes <- 28

# Number of leads and samples
num_leads <- 12
num_samples <- 5000

for (athlete in 1:num_athletes) {
  # Generate the file path based on the athlete number
  # Assuming file names are consistently formatted
  file_path <- sprintf("/content/ath_%03d.dat", athlete)

  # Open the file in binary mode
  con <- file(file_path, "rb")

  # Initialize a matrix to hold the data for each lead
  ecg_data <- matrix(nrow = num_samples, ncol = num_leads)

  # Reading the data
  for (i in 1:num_samples) {
    for (j in 1:num_leads) {
      # Read 2 bytes (16 bits) and convert to integer
      ecg_data[i, j] <- readBin(con, what = "integer", size = 2, n = 1, signed = TRUE, endian = "little")
    }
  }

  # Close the connection
  close(con)

  # Store the data in the list
  ecg_data_list[[athlete]] <- ecg_data
}

# At this point, ecg_data_list contains the ECG data for all 28 athletes
# You can access each athlete's data using ecg_data_list[[athlete_number]]
```

```r
ecg_data_df_list <- lapply(ecg_data_list, function(ecg_matrix) {
  colnames(ecg_matrix) <- paste0("V", 1:num_leads)
  return(as.data.frame(ecg_matrix))
})
```

```r
ecg_data_df_list[1]
```

```r
# Initialize an empty data frame to store the combined data
combined_ecg_data_df <- data.frame()

# Loop through the list of ECG data matrices
for (athlete in 1:length(ecg_data_list)) {
  # Convert the current matrix to a data frame
  ecg_data_df <- as.data.frame(ecg_data_list[[athlete]])

  # Assign column names V1 to V12
  colnames(ecg_data_df) <- paste0("V", 1:num_leads)

  # Add an athlete identifier column
  ecg_data_df$Athlete <- rep(athlete, nrow(ecg_data_df))

  # Combine the current athlete's data frame with the main data frame
  combined_ecg_data_df <- rbind(combined_ecg_data_df, ecg_data_df)
}

# Now, combined_ecg_data_df contains the data for all athletes with proper column names
# and an additional column for the athlete identifier.
```

```r
for (athlete in 1:28) {
  # Filter data for the current athlete and the first 500 samples
  ecg_subset <- combined_ecg_data_df %>%
    filter(Athlete == athlete) %>%
    head(500) # Adjust the number of samples if you want to visualize a longer segment

  # Generate the plot
  p <- ggplot(ecg_subset, aes(x = 1:500, y = V1)) +
    geom_line() +
    labs(title = paste("ECG Waveform for Athlete", athlete, ", Lead V1"),
         x = "Sample Index",
         y = "Amplitude") +
    theme_minimal()

  # Print the plot to the R graphics device
  print(p)
}
```
